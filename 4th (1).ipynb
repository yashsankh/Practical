Title : Implement Gradient Descent Algorithm to find the local minima of a function.
For example, find the local minima of the function y=(x+3)Â² starting from the point x=2.

# Gradient Descent Algorithm

```python
x=2 #starting point
lr=0.01 #learning rate
precision=0.000001
previous_step_size=1
iter=0
max_iter=10000
gf=lambda x: (x+3)**2 # function
```

```python
import matplotlib.pyplot as plt
gd=[]
```

```python
while precision < previous_step_size and iter < max_iter:
    prev=x
    x=x-lr*gf(prev)
    previous_step_size=abs(x-prev)
    iter+=1
    print('iteration:',iter,'Value:',x)
    gd.append(x)
```

```python
print('Local Minima',x)
```

# Gradient Descent

```python
plt.plot(gd)
```
